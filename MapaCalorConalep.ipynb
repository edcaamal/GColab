```python
# Librerias
import folium
from folium.plugins import MarkerCluster
from folium.plugins import HeatMap
import pandas as pd
```

```python
loc = pd.read_excel("/content/clasefolium.xlsx")
loc
```

```python
Mapa = folium.Map(location=[loc['lon'].mean(),loc['lat'].mean()],
                  zoom_start=16)
marker_cluster = MarkerCluster().add_to(Mapa)
for i in range (0, len(loc)):
    folium.Marker([loc.iloc[i]['lon'], loc.iloc[i]['lat']],
              popup=loc.iloc[i]['nombre'],
              icon=folium.map.Icon(color=loc.iloc[i]['color'],
                      icon=loc.iloc[i]['icono'],prefix='fa')).add_to(marker_cluster)

Mapa
```

```python
locais = []
locais = loc[['lon', 'lat']].values.tolist()
HeatMap(locais, radius=20).add_to(Mapa)
Mapa
```
